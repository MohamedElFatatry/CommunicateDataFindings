## Trip Counts per Day & Hour

> Do you think More Trips are done in specific days more than another ?

>What do you think are the rush hours during the day ?

```python
fig=plt.figure(figsize=(8, 6));
order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sb.countplot(data=df_clean, x='start_day', order = order);
plt.xlabel('Trip Day',fontsize=25);
plt.xticks(rotation=90 , fontsize=15);
plt.yticks(fontsize=15);
plt.ylabel('Count',fontsize=25);
plt.title('Trip Counts per Day', fontsize=30);
```

```python
plt.figure(figsize=(12, 6))
sb.countplot(data = df_clean, x='start_hour')
plt.xlabel('Trip Start hour',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.title('Trip counts per hour',fontsize=25);
```

### Observation : 

>As we can notice 'Thursday' is the day with highest number of trips.

>we can also notice number of trips significally decrease in weekends ( saturday & sunday)

>As we can see the two hours with most number of trips are 8 AM (the begining of the working day) & 5 PM (the end of the working day(rush hour))

>As noticed number of trips starts decreasing gradually starting from 6 PM and starts increasing again at 4 AM (least activity during sleeping hours)

## Trip Durations With Age, Gender & User Type

>Is there Correlation between trip duration & age ?

>What gender take longer trips ?

>who takes longer trips ? Subscribers or Costumers ?

```python
plt.figure(figsize=[20,5])
plt.subplot(1, 2, 1)
plt.scatter(df_clean['age'], df_clean['duration_min'], alpha = 0.35)
plt.axis([15, 100,0, 700])
plt.title('Trip Durations compared to Age', fontsize=30)
plt.xlabel('Age', fontsize=25)
plt.ylabel('Duaration per Minute', fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15);
```

```python
plt.figure(figsize = [20, 10])
sb.boxplot(data = df_clean, x = 'member_gender', y = 'duration_min')
plt.ylim([1, 40])
plt.title('Trip Duration and Gender',fontsize=35)
plt.xlabel('Gender',fontsize=25)
plt.ylabel('Duration per Minute', fontsize=25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show();
```

```python
plt.figure(figsize = [10, 8])
sb.boxplot(data = df_clean, x = 'user_type', y = 'duration_min')
plt.ylim([0, 50])
plt.title('Trip Duration and User Type', fontsize=25)
plt.xlabel('User Type', fontsize=20)
plt.ylabel('Duration in Minutes', fontsize=20)
plt.show()
```

### Observation :
>It is obvious that trip durations decrease as age increase

>Mean trip durations for females and other is higher than that of males.

>Higher percentage of customers' trips have longer durations than these of subscribers
